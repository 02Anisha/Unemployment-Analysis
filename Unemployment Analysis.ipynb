```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import plotly.express as ex
import seaborn as sns
```

# Read the csv file

```python
df = pd.read_csv("Unemployment_Rate.csv")
```

## Check if there are any missing values

```python
print(df.isnull().sum())
```

```python
#Rename colomns
df.columns =['States','Date','Frequency','Estimated Unemployment Rate(%)','Estimated Employed','Estimated Labour Participation Rate(%)','Region','Longitude','Latitude']
```

## **Visualizing the relatio**n

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots()
sns.heatmap(df.corr())
plt.show()
```

```python
#Look at the estimated employees in different regions of India
df.columns=['States','Date','Frequency','Estimated Unemployment Rate(%)','Estimated Employed','Estimated Labour Participation Rate(%)','Region','Longitude','Latitude']
plt.title('Unemployment in India')
sns.histplot(x ="Estimated Employed",hue ="Region",data = df)
plt.show()

                                          
                   
                                     
```

```python
# analyze the unemployment rate in different regions in India
plt.figure(figsize =(12,10))
plt.title("Unemployment in India")
sns.histplot(x ="Estimated Unemployment Rate(%)",hue ="Region",data = df)
plt.show()


```

```python
unemployment = df[["States","Region","Estimated Unemployment Rate(%)"]]
figure =ex.sunburst(unemployment, path =["Region","States"],
                    values ="Estimated Unemployment Rate(%)",width =500,height =500,
                    color_continuous_scale ="RdY1Gn",title ="Unemployment Rate in India")
figure.show()
```
